## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
array = np.random.randint(0, 1000, size=100, dtype=np.int32)

print("Mảng ngẫu nhiên gồm 100 phần tử:")
print(array)

print("\n10 phần tử đầu tiên của mảng:")
print(array[:10])

print("\n5 phần tử cuối cùng của mảng:")
print(array[-5:])

print("\nCác phần tử chia hết cho 2 và 5:")
print(array[(array % 2 == 0) & (array % 5 == 0)])
```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
print(array[array < 100].size)
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
print("Tổng mảng: ")
print(np.sum(array))

print("Giá trị trung bình: ")
print(np.mean(array))

print("Tổng mảng: ")
print(np.median(array))

print("Q1:  ")
print(np.percentile(array, 25))
print("Q1:  ")
print(np.percentile(array, 50))
print("Q1:  ")
print(np.percentile(array, 70))

print("Phân vị 20:  ")
print(np.percentile(array, 20))
```

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

```python
for i in range(1, 11):
  chosen = np.random.choice(array, size=10, replace=False)
  print("\nLần " + str(i))
  print(chosen)
```

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

```python
values, counts = np.unique(array, return_counts=True)

most_frequency_value = np.max(counts)
indices = np.where(counts == most_frequency_value)[0]

for i in range(1, len(indices)):
  print(values[i], end=" ")
  if (i % 10 == 0):
    print()

```

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

```python
# YOUR CODE HERE
new_array_2 = np.linspace(0, np.pi, 1000)

y1 = np.sin(new_array_2)
y2 = np.cos(new_array_2)


plt.plot(new_array_2, y1)
plt.plot(new_array_2, y2, linestyle='--')
plt.show()
```

### Câu 5: Vẽ các đồ thị con

```python
plt.subplot(2,1,1)
plt.plot(new_array_2, y1)

plt.subplot(2,1,2)
plt.plot(new_array_2, y2, linestyle='--')

plt.show()
```

### Câu 8: Vẽ histogram

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ histogram các số ngẫu nhiên trên.

```python
# YOUR CODE HERE
x = np.random.normal(size=10000)     

plt.hist(x, bins=100)
plt.xlabel('Giá trị')
plt.ylabel('Tần số')
plt.title('Histogram')

plt.show()
```

### Câu 9: Vẽ đồ thị hộp

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ boxplot các số ngẫu nhiên trên.

```python
x = np.random.normal(size=10000)  

plt.boxplot(x=x)
plt.title("Boxplot")

plt.show()
```
